```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from your dataset (replace 'data.csv' with your actual file path)
data = pd.read_csv('final_table_E_new.csv',encoding='latin1')

# Create a bar plot
plt.figure(figsize=(8, 6))
data.groupby('type')['Degree'].mean().plot(kind='bar', color=['blue', 'green'])

# Set labels and title
plt.xlabel('Game Type')
plt.ylabel('Mean Degree')
plt.title('Mean Degree by Game Type')

# Show plot
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from your dataset (replace 'data.csv' with your actual file path)
data = pd.read_csv('final_table_E_new.csv',encoding='latin1')

# Filter data for 'title'
title_data = data[data['type'] == 'Title']

# Create an area plot for 'title' vs. 'eccentricity'
plt.figure(figsize=(10, 6))
plt.fill_between(title_data.index, title_data['Eccentricity'], color='skyblue', alpha=0.4)

# Set labels and title
plt.xlabel('Index')
plt.ylabel('Eccentricity Value')
plt.title('Area Plot of Title vs. Eccentricity')

# Show plot
plt.tight_layout()
plt.show()
```

```python
team_data = data[data['type'] == 'Team']

# Create an area plot for 'team' vs. 'eccentricity'
plt.figure(figsize=(10, 6))
plt.fill_between(team_data.index, team_data['Eccentricity'], color='skyblue', alpha=0.4)

# Set labels and title
plt.xlabel('Index')
plt.ylabel('Eccentricity Value')
plt.title('Area Plot of Team vs. Eccentricity')

# Show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read your dataset
# Assuming your dataset is stored in a variable named 'df'

# Group the dataset by the 'type' column
grouped = data.groupby('type')

# Define the numerical columns of interest
numerical_columns = ['Degree', 'Weighted Degree', 'Eccentricity', 'Closeness Centrality',
                     'Harmonic Centrality', 'Betweeness Centrality', 'Authority', 'Hub',
                     'Eigenvector centrality', 'clustering coefficient', 'Number of triangles',
                     'PageRank', 'Component Number', 'Modularity']

# Create box plots for each numerical column for each group
for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    positions = range(len(grouped))
    for position, (group_name, group_data) in zip(positions, grouped):
        plt.boxplot(group_data[column], positions=[position], widths=0.5, patch_artist=True)
    plt.title(f'Box Plot of {column} by Type')
    plt.xlabel('Type')
    plt.ylabel(column)
    plt.xticks(positions, [group_name for group_name, _ in grouped])
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read your dataset
# Assuming your dataset is stored in a variable named 'df'

# Select the columns for which you want to compute correlations
numerical_columns = ['Degree', 'Weighted Degree', 'Eccentricity', 'Closeness Centrality',
                     'Harmonic Centrality', 'Betweeness Centrality', 'Authority', 'Hub',
                     'Eigenvector centrality', 'clustering coefficient', 'Number of triangles',
                     'PageRank', 'Component Number', 'Modularity']

# Filter the dataset to include only the numerical columns
numerical_data = data[numerical_columns]

# Compute the correlation matrix
correlation_matrix = numerical_data.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
# import pandas as pd
# import matplotlib.pyplot as plt

# # Read your dataset
# # Assuming your dataset is stored in a variable named 'df'

# # Select the columns for the plot
# numerical_columns = ['Degree', 'Weighted Degree', 'Eccentricity', 'Closeness Centrality',
#                      'Harmonic Centrality', 'Betweeness Centrality', 'Authority', 'Hub',
#                      'Eigenvector centrality', 'clustering coefficient', 'Number of triangles',
#                      'PageRank', 'Component Number', 'Modularity']

# # Filter the dataset to include only the numerical columns
# numerical_data = data[numerical_columns]

# # Normalize the data (optional but recommended)
# normalized_data = (numerical_data - numerical_data.min()) / (numerical_data.max() - numerical_data.min())

# # Create a temporary class column
# normalized_data['class_column'] = range(len(normalized_data))

# # Plot the Parallel Coordinates Plot
# plt.figure(figsize=(12, 8))
# pd.plotting.parallel_coordinates(normalized_data, 'class_column', color='blue', alpha=0.4)
# plt.title('Parallel Coordinates Plot of Numerical Variables')
# plt.xlabel('Variables')
# plt.ylabel('Normalized Values')
# plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
# plt.tight_layout()  # Adjust layout to prevent overlap
# plt.show()

# import pandas as pd
# import matplotlib.pyplot as plt
# from matplotlib.colors import Normalize
# from matplotlib.cm import get_cmap

# # Read your dataset
# # Assuming your dataset is stored in a variable named 'df'

# # Select the columns for the plot
# numerical_columns = ['Degree', 'Weighted Degree', 'Eccentricity', 'Closeness Centrality',
#                      'Harmonic Centrality', 'Betweeness Centrality', 'Authority', 'Hub',
#                      'Eigenvector centrality', 'clustering coefficient', 'Number of triangles',
#                      'PageRank', 'Component Number', 'Modularity']

# # Filter the dataset to include only the numerical columns
# numerical_data = data[numerical_columns]

# # Normalize the data (optional but recommended)
# normalized_data = (numerical_data - numerical_data.min()) / (numerical_data.max() - numerical_data.min())

# # Assign a color to each row
# norm = Normalize(vmin=0, vmax=len(normalized_data)-1)
# cmap = get_cmap('viridis')

# # Plot the Parallel Coordinates Plot
# plt.figure(figsize=(12, 8))
# for i in range(len(normalized_data)):
#     color = cmap(norm(i))
#     plt.plot(normalized_data.iloc[i], color=color, alpha=0.4)

# plt.title('Parallel Coordinates Plot of Numerical Variables')
# plt.xlabel('Variables')
# plt.ylabel('Normalized Values')
# plt.xticks(range(len(numerical_columns)), numerical_columns, rotation=45)
# plt.legend(loc='upper right')
# plt.tight_layout()  # Adjust layout to prevent overlap
# plt.show()


import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from matplotlib.cm import get_cmap

# Read your dataset
# Assuming your dataset is stored in a variable named 'df'

# Select the columns for the plot
numerical_columns = ['Degree', 'Weighted Degree', 'Eccentricity', 'Closeness Centrality',
                     'Harmonic Centrality', 'Betweeness Centrality', 'Authority', 'Hub',
                     'Eigenvector centrality', 'clustering coefficient', 'Number of triangles',
                     'PageRank', 'Component Number', 'Modularity']

# Filter the dataset to include only the numerical columns
numerical_data = data[numerical_columns]

# Normalize the data (optional but recommended)
normalized_data = (numerical_data - numerical_data.min()) / (numerical_data.max() - numerical_data.min())

# Assign a color to each row
norm = Normalize(vmin=0, vmax=len(normalized_data)-1)
cmap = get_cmap('viridis')

# Plot the Parallel Coordinates Plot
plt.figure(figsize=(12, 8))
for i in range(len(normalized_data)):
    color = cmap(norm(i))
    plt.plot(normalized_data.columns, normalized_data.iloc[i], color=color, alpha=0.4, label=f'Row {i+1}')

plt.title('Parallel Coordinates Plot of Numerical Variables')
plt.xlabel('Variables')
plt.ylabel('Normalized Values')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
